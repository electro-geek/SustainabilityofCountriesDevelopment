```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
dataset = pd.read_csv('sed.csv')
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, columns = ['region', 'level_of_development'], drop_first = True)
```

```python
dataset.head()
```

```python
for i in dataset['dt_date']:
    y = i.split("-")
    temp = int(y[0]+y[1]+y[2])
    dataset = dataset.replace(i,temp)
```

```python
dataset.head()
```

```python
#to check which column contain what type of value
k = 0
count = 0
for i in dataset.columns:
    for j in dataset[i]:
        print(type(j))
        k += 1
        count += 1
        if k > 0:
            break
        
print(count)
print(len(dataset.columns))
```

```python
list1 = dataset.columns
print(len(list1))
```

```python
#handling string column of the country
f = 0
dict1 = dict()
for i in dataset['country']:
    if i not in dict1:
        dict1[i] = f
        f += 1
```

```python
dict1
```

```python
for i in dataset['country']:
    dataset = dataset.replace(i,dict1[i])
```
